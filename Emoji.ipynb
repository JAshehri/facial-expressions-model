```python
# csv analysis 
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split

#zip file
import zipfile
import os

#cnn model
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator


# img
import matplotlib.pyplot as plt
import cv2
```

```python
#ex zip file
zpath =r"C:\Users\JoudA\ML\archive.zip"
expath= r"C:\Users\JoudA\ML\extracted"   

with zipfile.ZipFile(zpath,"r") as archive:
    archive.extractall(expath)

print ("done",expath)
```

```python
modle = Sequential ([
#config 3 cnn layer 
    #layern 1 with 32 filter
    Conv2D(32 ,(3,3), activation='relu',input_shape=(48,48 ,1)),
    MaxPooling2D((2,2)),

    #layern 2 with 64 filter
    Conv2D(64 ,(3,3), activation='relu'),
    MaxPooling2D((2,2)),

    #layern 3 with 128 filter
    Conv2D(128,(3,3), activation='relu'),
    MaxPooling2D((2,2)),

    #layern 4 with 128 filter
    Conv2D(256,(3,3), activation='relu'),
    MaxPooling2D((2,2)),

    #from 1D ot 2D
    Flatten(),

    #hidden alyer with 128 nerves 
    Dense(256, activation='relu'),
   
    #Overfitting solving
    Dropout(0.5),

    #last layer (7 output)
    Dense(7, activation='softmax')
])
```

```python
# resize pectuers

td=ImageDataGenerator(
    rescale=1./255,
)
```

```python
#pectuer path 
train_path =r"C:\Users\JoudA\ML\extracted\train"
test_path=r"C:\Users\JoudA\ML\extracted\test"

#upload trainning data
tr_g=td.flow_from_directory(
    train_path,
    target_size=(48,48),
    color_mode='grayscale',
    batch_size=32,
    class_mode='categorical'
)

#upload test data
ts_g=td.flow_from_directory(
    test_path,
    target_size=(48,48),
    color_mode='grayscale',
    batch_size=32,
    class_mode='categorical'
)
```

```python
# model compail
modle.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

#model trainning
history= modle.fit(
    tr_g,
    epochs=50,
    validation_data=tr_g
)
```
